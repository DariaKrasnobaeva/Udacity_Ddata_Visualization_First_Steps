**Task 1**: Let's look at the relationship between fuel mileage ratings for city vs. highway driving, as stored in the 'city' and 'highway' variables (in miles per gallon, or mpg). Use a _scatter plot_ to depict the data. What is the general relationship between these variables? Are there any points that appear unusual against these trends?

```python
# YOUR CODE HERE
plt.scatter(data = fuel_econ, x ='city', y='highway');
plt.xlabel('mileage ratings for city')
plt.ylabel('mileage ratings for highway')
```

```python
# YOUR CODE HERE with seaborn
sb.regplot(data = fuel_econ, x ='city', y='highway');
plt.xlabel('mileage ratings for city')
plt.ylabel('mileage ratings for highway');
```

```python
#without regression line
sb.regplot(data = fuel_econ, x ='city', y='highway', fit_reg=False);
plt.xlabel('mileage ratings for city')
plt.ylabel('mileage ratings for highway');
```

```python
#without regression line
sb.regplot(data = fuel_econ, x ='city', y='highway', x_jitter= 0.3,
          scatter_kws={'alpha':1/20}); #adding jitter
plt.xlabel('mileage ratings for city')
plt.ylabel('mileage ratings for highway');
```

**Task 2**: Let's look at the relationship between two other numeric variables. How does the engine size relate to a car's CO2 footprint? The 'displ' variable has the former (in liters), while the 'co2' variable has the latter (in grams per mile). Use a heat map to depict the data. How strong is this trend?

```python
# YOUR CODE HERE
plt.hist2d(data = fuel_econ, x ='displ', y='co2'); 
plt.colorbar()
plt.xlabel('engine size')
plt.ylabel('CO2 footprint');
```

```python
plt.hist2d(data = fuel_econ, x ='displ', y='co2', cmin=0.5); #changing background
plt.colorbar()
plt.xlabel('engine size')
plt.ylabel('CO2 footprint');
```

```python
plt.hist2d(data = fuel_econ, x ='displ', y='co2', cmin=0.5, cmap='viridis_r'); #adding reverse color scheme
plt.colorbar()
plt.xlabel('engine size')
plt.ylabel('CO2 footprint');
```

```python
fuel_econ[['displ','co2']].describe()
```

```python
bins_x=np.arange(0.6, 7+0.3,0.3)
bins_y=np.arange(29, 692+30,30) #information from describe
plt.hist2d(data = fuel_econ, x ='displ', y='co2', cmin=0.5, cmap='viridis_r', bins =[bins_x, bins_y]); #adding bins
plt.colorbar()
plt.xlabel('engine size')
plt.ylabel('CO2 footprint');
```
